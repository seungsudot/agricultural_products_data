```python
# 각 세대를 기준으로 분기별 판매량 합치는 코드 

year = 2023

# 각각의 프레임을 불러온다. 
for age in range (2,9):
    df_01 = pd.read_csv(f"./example_file/2023_01_{age*10}_age_group.csv", index_col =0)
    df_02 = pd.read_csv(f"./example_file/2023_02_{age*10}_age_group.csv", index_col =0)
    df_03 = pd.read_csv(f"./example_file/2023_03_{age*10}_age_group.csv", index_col =0)
    df_04 = pd.read_csv(f"./example_file/2023_04_{age*10}_age_group.csv", index_col =0)
    
    # value_counts형태로 바꿔서 소비량을 한눈에 보이게 만든다.
    val_count_01 = df_01["wide"].value_counts()
    val_count_02 = df_02["wide"].value_counts()
    val_count_03 = df_03["wide"].value_counts()
    val_count_04 = df_04["wide"].value_counts()
    
    concat_df = pd.concat([val_count_01,val_count_02], ignore_index=True, axis = 1)
    concat_df = pd.concat([concat_df,val_count_03], ignore_index=True, axis = 1)
    concat_df = pd.concat([concat_df,val_count_04], ignore_index=True, axis = 1)
    
    concat_df.to_csv(f"./example_file/concat/concat_resut_{age*10}_age_group.csv")
    concat_df.to_excel(f"./example_file/concat/concat_resut_{age*10}_age_group.xlsx")
    
    
```
